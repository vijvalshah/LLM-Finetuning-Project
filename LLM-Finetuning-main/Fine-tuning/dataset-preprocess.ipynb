```python
with open(test_data_file_path, 'r') as f:
    line = f.readline()
    data = json.loads(line)

# comment out the following lines to process for review comment generation task


# patch = process_diff(data['patch'])
# print(patch)
# msg = data['msg']
# print(msg)


# lines to process for code refinement generation task

oldlines, newlines, comment = process_input(data)
print(oldlines)
print(newlines)
print(comment)
```

```python
# for review comment generation task
# prompt = """You are a powerful code reviewer model. Your job is to suggest review comment in natural language. You are given a question, and context regarding a diff hunk or code change in programming language. You must output appropriate, contextual review comment for that code change."""

# for code refinement generation task
prompt = """You are a powerful code reviewer model. Your job is to suggest refined or fixed code based on the natural language review comment. You are given a question, and context regarding an old diff hunk or code change in programming language. You are also given a review comment based on that old code. You must output accurate refined, fixed new code snippet for that old code change and corresponding review comment in the same programming language as the old code. """
```

```python
def read_jsonl(path):
    data = []
    with open(path) as f:
        for line in f:
            try:
                js = json.loads(line.strip())
            except:
                print("Error during reading json data.")
                continue
            data.append(js)
    return data
```

```python
# train_data = read_jsonl(train_data_file_path)
test_data = read_jsonl(test_data_file_path)

# for i in range(len(train_data)):
#     train_data[i]["idx"] = i
for i in range(len(test_data)):
    test_data[i]["idx"] = i


# comment out the following lines to process for review comment generation task

# for dic in train_data:
#     dic["input"] = process_diff(dic["patch"])
#     dic["instruction"] = prompt
#     dic["output"] = dic["msg"]
# for dic in test_data:
#     dic["input"] = process_diff(dic["patch"])
#     dic["instruction"] = prompt
#     dic["output"] = dic["msg"]


# lines to process for code refinement generation task

# for dic in train_data:
#     old, new, comment = process_input(dic)
#     dic["input"] = "Review Comment: " + comment + "\nOld Code: " + old 
#     dic["instruction"] = prompt
#     dic["output"] = "New Code: " + new
for dic in test_data:
    old, new, comment, _, _ = process_input(dic)
    dic["input"] = "Review Comment: " + comment + "\nOld Code: " + old 
    dic["instruction"] = prompt
    dic["output"] = "New Code: " + new
```

```python
# load the data into pandas dataframe in two columns named "diff" and "msg"
import pandas as pd
test_df = pd.DataFrame(test_data)
test_df = test_df[["instruction", "input", "output"]]
# df.iloc[0]["diff"]
test_df
```
